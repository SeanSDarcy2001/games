```python
class Player() :

  def __init__(self, Name) :
    self.name = Name
    self.hand = []
    self.chips = 0

  def buyIn(self, amnt) :
    self.chips = amnt

  def bet(self, amnt) :
    amnt = int(amnt)
    if amnt <= self.chips :
      self.chips -= amnt
      betPlaced = False
      return amnt
    else :
      print(self.getName(), "is all in.")
      allIn = self.chips
      self.chips -= allIn
      return allIn

  def collect(self, amnt) :
    self.chips += amnt
  
  def amIup(self) :
    return self.chips
  
  def getName(self) :
    return self.name

  def updateHand(self, card) :
    self.hand.append(card)
  
  def getHand(self) :
    value = 0
    numAces = 0
    cards = []
    for i in range(len(self.hand)) :
      cards.append(self.hand[i].getCard())
      value += self.hand[i].value()

      if self.hand[i].card == "Ace" :
        numAces += 1
      
    while(value > 21 and numAces > 0) :
      value -= 10
      numAces -= 1
    return cards, value

  def resetHand(self) :
    self.hand = []

    
class Card() :

  def __init__(self, suit, card) :
    self.suit = suit
    self.card = card

  def getCard(self) :
    return self.card + " of " + self.suit
  
  def value(self) :
    if self.card == "Ace" :
      val = 11
    elif self.card == "2" :
      val = 2
    elif self.card == "3" :
      val = 3
    elif self.card == "4" :
      val = 4
    elif self.card == "5" :
      val = 5
    elif self.card == "6" :
      val = 6
    elif self.card == "7" :
      val = 7
    elif self.card == "8" :
      val = 8
    elif self.card == "9" :
      val = 9
    elif self.card == "10" or self.card == "Jack" or self.card == "Queen" or self.card == "King":
      val = 10
    return val
    
class Deck() :
   def __init__(self) :
     
     self.suits = ["Spades", "Diamonds", "Clubs", "Hearts"]
     self.cards = ["Ace", "2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King"]
     self.deck = []

     for suit in self.suits :
       for card in self.cards :
         self.deck.append(Card(suit, card))
    
   def shuffleDeck(self) :
     random.shuffle(self.deck)
  
   def draw(self) :
     return self.deck.pop()

   def cardsLeft(self) :
     return len(self.deck)


class Dealer() :
  def __init__(self, players) :
    self.deck = Deck()
    self.table = players
    self.hand = []
    
  def shuffle(self):
    self.deck.shuffleDeck()
  
  def deal(self) :
    for i in range(2) :
      if self.checkDeck() < len(self.table) + 1:
        self.deck = Deck()
        self.shuffle()
        print("New Deck!!!")
      for players in self.table :
        players.updateHand(self.deck.draw())
      self.hand.append(self.deck.draw())

  def whatsUp(self) :
    return self.hand[1].getCard()

  def whatsDown(self) :
    return self.hand[0].getCard()

  def hit(self, player) :
    if self.checkDeck() < 1 :
        self.deck = Deck()
        self.shuffle()
        print("New Deck!!!")
    player.updateHand(self.deck.draw())
  
  def pay(self, player, amnt) :
    player.collect(amnt)

  def getHand(self) :
    value = 0
    numAces = 0
    cards = []
    for i in range(len(self.hand)) :
      cards.append(self.hand[i].getCard())
      value += self.hand[i].value()

      if self.hand[i].card == "Ace" :
        numAces += 1
      
    while(value > 21 and numAces > 0) :
      value -= 10
      numAces -= 1
    return cards, value

  def dealerHit(self) :
    self.hand.append(self.deck.draw())
  
  def resetHand(self) :
    self.hand = []

  def checkDeck(self) :
    return self.deck.cardsLeft()

```
